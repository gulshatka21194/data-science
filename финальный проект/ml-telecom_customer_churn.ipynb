Необходимо объединить данные для удобства изучения данных и выполнения дальнейшей обработки и анализа.

Объединим таблицы по столбцу customerID. Нужно сперва объединить все данные по звонкам и интернет-трафику, затем добавить инфу о контрактах и пользователях. Используем при объединении звонков и трафика outer join, чтобы не пропустить клиентов, использующих не все услуги. При объединении договоров также используем outer join, потому что есть клиенты, подключившие либо интернет, либо стационарную телефонную связь. А пользователей объединяем при помощи left, потому что количество договоров соответствует количеству клиентов. 

```python
print(df_phone.shape)
data = df_phone.merge(df_internet, on='customerID', how='outer')
print(data.shape)
data = data.merge(df_contract, on='customerID', how='outer')
print(data.shape)
data = data.merge(df_personal, on='customerID', how='left')
print(data.shape)
```

### Изменение типов данных

Заменим в столбце begin_date тип object на datetime64.

```python
data['begin_date'] = pd.to_datetime(data['begin_date'], format='%Y-%m-%d') #заменим тип object на datetime64
```

Заменим в столбце total_charges тип object на float64.

```python
data['total_charges'] = pd.to_numeric(data['total_charges'], errors='coerce') #заменим тип object на float64
```

Заменим в столбце senior_citizen тип Int на object.

```python
data['senior_citizen'] = data['senior_citizen'].apply(lambda x: 'No' if x == 0 else 'Yes') #применим lambda-выражение(условие: если значение равно 0, то присваиваем 'No', иначе 'Yes') к senior_citizen
```

```python
data['senior_citizen'] = data['senior_citizen'].astype('object') #заменим тип int64 на object
```

### Заполнение пропусков

В столбцах 'internet_service', 'online_security', 'online_backup', 'device_protection', 'tech_support', 'streaming_tv', 'streaming_movies', 'multiple_lines' есть пропущенные значения. Так как в этих столбцах представлены услуги, то отсутствие значений в них после объединения в одну таблицу означает, что клиенты просто не воспользовались ими.
```python
cols = ['internet_service', 'online_security', 'online_backup', 'device_protection',
        'tech_support', 'streaming_tv', 'streaming_movies', 'multiple_lines'] #определим список столбцов с пропущенными значениями
```

```python
for col in cols:
    data[col] = data[col].fillna('No') #в цикле заполним пустые значения 'No'
```

В столбце total_charges пропущенные значения говорят о том, что услуга была подключена в день сбора информации, т.е. 1 февраля 2020 года. Можем удалить эти данные.

```python
data.loc[data['begin_date'] == '2020-02-01']
```

```python
data = data.dropna(subset=['total_charges']) #удалим
```

Проверим.

### Расчет новых признаков

Определим целевой признак - факт ухода клиента, исходя из даты окончания договора.

```python
data['target'] = data['end_date'].apply(lambda x: 0 if x == 'No' else 1) #применим lambda-выражение(условие: если значение равно 'No', то присваиваем 0, иначе 1) к end_date
```

Также определим еще один новый признак - срок пользования клиентом услугами.

Для этого сперва заменим тип object на datetime64 и напишем функцию, в которой вычислим длительность вычитая из значения столбца end_date значение столбца begin_date.

```python
data['end_date'] = data['end_date'].apply(lambda x: '2020-02-01' if x == 'No' else x) #применим lambda-выражение(условие: если значение равно 'No', то присваиваем '2020-02-01', иначе оставляем значение) к end_date
```

```python
data['end_date'] = pd.to_datetime(data['end_date'], format='%Y-%m-%d') #заменим тип object на datetime64
```

```python
def calc_duration(row):
    return (row['end_date'] - row['begin_date']).days

data['duration'] = data.apply(calc_duration, axis=1)
```

Удалим столбец customer_id, так как является идентификатором, он будет сильно коррелировать с остальными столбцами.

```python
data = data.drop(['customer_id'], axis=1) # #удалим столбец 'customer_id'
```

Проверим данные на зависимость, чтобы в случае высокой связи исключить какой-либо признак.

```python
data.phik_matrix()
```

## Обучение моделей

Напишем функцию для расчета времени выполнения подбора гиперпараметров.

```python
def getTimeModel(grid, features_train, target_train):
    start = time.time() #получим текущее время и запишем в переменную start
    grid.fit(features_train, target_train) #выполним обучение модели
    predictions = grid.predict(features_train) #получим предсказания модели 
    stop = time.time() #получим время после выполнения обучения и запишем в переменную stop
    total_time = stop - start #из stop вычтем start
    print('Время выполнения обучения модели', total_time) #выведем время выполнения обучения модели
    print('Лучшее значение метрики roc-auc:', grid.best_score_) #выведем наилучший показатель
    print('Лучшие параметры:', grid.best_params_) #выведем наилучшие параметры
    return total_time, grid.best_score_, grid.best_params_, grid.best_estimator_.feature_importances_, predictions #вернем время, метрику, наилучшие параметры, важность признаков и предсказания
```

```python
data_result = []
```

```python
parametres = {'max_depth': range(1, 12, 2)} #определим гиперпараметры


model_dt = DecisionTreeClassifier(random_state=190922, class_weight='balanced') #обучим модель с заданной глубиной дерева
search_cv = GridSearchCV(model_dt, parametres, scoring='roc_auc', cv=5) #применим GridSearchCV для модели model_dt
total_time, auc_roc, best_params, feat_imp, preds = getTimeModel(search_cv, features_train, target_train) #применим функцию getTimeModel
```

Добавим полученные результаты в таблицу.

```python
data_result.append(['DecisionTreeClassifier', total_time, auc_roc, best_params, feat_imp, preds]) #добавим total_time, auc_roc, best_params, feat_imp, preds в data_result
```
