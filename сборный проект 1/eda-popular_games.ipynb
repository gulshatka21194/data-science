## Предобработка данных

### Приведение столбцов к нижнему регистру

Приведем названия столбцов к змеиному регистру.

```python
data.columns = data.columns.str.lower()
```

### Замена пропусков

В столбцах name, genre есть пропущенные значения. Целесообразно их удалить.

```python
data = data.dropna(subset=['name', 'genre'])
```

Для рейтингов заменить пропуски на константу может быть логично, т.к. не во всех регионах эти рейтинги используются.

```python
data['rating'] = data['rating'].fillna('unknown')
```

В столбце user_score обнаружено значение 'tbd'. Что значит - подлежит обсуждению. Заменим это значение на пропуск.

```python
data.loc[data['user_score']=='tbd', 'user_score']=np.nan
```

```python
data
```

### Изменение типов данных

В столбце user_score заменим тип object на float64 и умножим на 10, для того, чтобы значения оценок пользователей и критиков были в одном интервале.

```python
data['user_score'] = pd.to_numeric(data['user_score'], errors='coerce') * 10 #заменим тип object на float64 и умножим на 10
```

Затем в столбцах year_of_release, user_score, critic_score заменим тип float64 на int64.

```python
data_columns = ['year_of_release', 'critic_score', 'user_score']

for val in data_columns:
    data[val] = data[val].astype('Int64')
```

Проверим.

```python
data.info()
```

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

### Посмотрите, сколько игр выпускалось в разные годы. Важны ли данные за все периоды?

Для того, чтобы посмотреть динамику выпуска игр в разные годы, сгруппируем таблицу по году выпуска, посчитаем количество по наименованию игр и построим график.

### Посмотрите, как менялись продажи по платформам. Выберите платформы с наибольшими суммарными продажами и постройте распределение по годам. За какой характерный срок появляются новые и исчезают старые платформы?

Сперва найдем платформы с наибольшими суммарными продажами. Для этого сгруппируем данные по платформам, найдем суммарные продажи и отсортируем по убыванию значения больше 100.

```python
data_top_platform = data.pivot_table(index='platform', values='total_sales', aggfunc={'sum'}).sort_values(by='sum', ascending=False).query('sum >= 100')
data_top_platform
```

Затем с помощью метода isin() отфильтруем исходную таблицу по топовым платформам.

```python
top_platform = data.query('platform.isin(@data_top_platform.index)')
top_platform
```

Построим график продаж игр на популярных платформах по годам.

```python
top_platform.groupby(['platform', 'year_of_release'])['total_sales'].sum().unstack('platform').plot(kind='line', figsize=(16, 8))
plt.xlabel('Год выпуска')
plt.ylabel('Продажи, млн.копий')
plt.title('Продажи игр на популярных платформах')
plt.show()
```

#### Вывод

Получили перечень наиболее популярных платформ - PS2, X360, Wii, PS3, DS, PS, GBA, PSP, PS4, PC.

На графике видим периоды пика платформ, но жизненным циклом является период пика, когда уже платформа обрела популярность, но не началась смена на новую платформу (следующего поколения). Таким образом, примерно за 4 года, начиная с 2013-года, появляются новые и исчезают старые платформы. Также из-за того, что данные за 2016-ый год неполные, целесообразней взять за верхнюю границу 2015-ый год.

### Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

Для анализа выберем платформу PS4. Для этого сделаем срез по условию platform == "PS4" и сгруппируем таблицу по оценкам пользователей. Рассчитаем суммарные продажи.

```python
one_platform_user_score = data_actual.query('platform == "PS4"').groupby('user_score')['total_sales'].sum().reset_index()
one_platform_user_score
```

Сделаем срез по условию platform == "PS4" и сгруппируем таблицу по оценкам критиков. Рассчитаем суммарные продажи.

```python
one_platform_critic_score = data_actual.query('platform == "PS4"').groupby('critic_score')['total_sales'].sum().reset_index()
one_platform_critic_score
```

Построим диаграмму рассеивания для анализа зависимости оценок пользователей/продаж и зависимости оценок критиков/продаж. Затем найдем коррелляции.

```python
plt.figure(figsize=(15,7))

one_platform_critic_score.plot(kind='scatter', x='critic_score', y='total_sales',
                      title = 'Зависимость оценок критиков и продаж, PS4')
one_platform_user_score.plot(kind='scatter', x='user_score', y='total_sales',
                      title = 'Зависимость оценок пользователей и продаж, PS4')
print('Коррелляция оценок пользователей и продаж:', one_platform_user_score['user_score'].corr(one_platform_user_score['total_sales']))
print('Коррелляция оценок критиков и продаж:', one_platform_critic_score['critic_score'].corr(one_platform_critic_score['total_sales']))
```

#### Вывод

Проверили зависимость оценок пользователей на продажи. Чем больше оценка, тем больше продажи.Об этом также говорит положительная корреляция. Коэффициент составляет 0.38.

Проверили зависимость оценок критиков на продажи. Чем больше оценка, тем больше продажи. Об этом также говорит положительная корреляция. Коэффициент составляет 0.52.

Исходя из результатов, зависимость от оценок критиков выше, чем от оценок пользователей.

## Проверка гипотез

### Проверка гипотезы 'Средние пользовательские рейтинги платформ Xbox One и PC одинаковые'

Сформулируем нулевую гипотезу:

H0 - Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

Тогда в качестве альтернативной возьмем следующую двустороннюю гипотезу:

H1 - Средние пользовательские рейтинги платформ Xbox One и PC разные

Сделаем срез таблицы по user_score с условием platform == "XOne", найдем дисперсию и количество.

```python
data_actual.query('platform == "XOne"')['user_score'].var()
```

```python
data_actual.query('platform == "XOne"')['user_score'].count()
```

Сделаем также срез таблицы по user_score с условием platform == "PC", найдем дисперсию и количество.

```python
data_actual.query('platform == "PC"')['user_score'].var()
```

```python
data_actual.query('platform == "PC"')['user_score'].count()
```

Для проверки гипотезы используем метод ttest_ind. Так как дисперсии не равны и размеры выборок отличаюся, параметр equal_var выставим False.

```python
filtered_xone = data_actual.query('platform == "XOne"')['user_score'].dropna().reset_index(drop=True)
filtered_pc = data_actual.query('platform == "PC"')['user_score'].dropna().reset_index(drop=True)

print('Средние пользовательские рейтинги платформы XOne:', filtered_xone.mean())
print('Средние пользовательские рейтинги платформы PC:', filtered_pc.mean())

alpha = .05 # критический уровень статистической значимости

results = st.ttest_ind(filtered_xone, filtered_pc, equal_var=False)

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

#### Вывод

При выбранном уровне значимости в 5% получили значение p-value 28%. Таким образом, нулевая гипотеза о равенстве средних не отвергается: Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
