## Предобработка данных

### Исследование и заполнение пропусков в данных

#### Приведение столбцов к нижнему регистру

Приведем название столбца cityCenters_nearest к змеиному регистру.

```python
data = data.rename(columns = {"cityCenters_nearest": "city_centers_nearest"}) #переименуем столбец 'cityCenters_nearest' на 'city_centers_nearest'
```

#### Замена пропусков

В столбце is_apartment 88% пропущенных значений. Но т.к. апартаменты — это юридический статус помещения, заполнить какими-либо значениями, исходя из имеющихся данных, невозможно. Оставим пропуски.

Заполнение средними значениями геоданных (airports_nearest, cityCenters_nearest, parks_around3000, parks_nearest, ponds_around3000, ponds_nearest) не совсем логично, т.к. расстояние дома от парков/рек и пр. никак не зависит от технических характеристик квартиры.

Пропущенные значения высоты потолка оставим без изменений. Т.к. заполнение медианой может повлиять на анализ зависимостей.

Также пропуски в столбце days_exposition (количество дней публикации) заполнять не будем, т.к. скорее всего отсутствие значения говорит о том, что квартира еще не продана.

Остальные пропуски попробуем заполнить.

##### Замена пропусков в столбце 'balcony'

В столбце balcony 48% пропущенных значений. Вероятно, если человек не указал число балконов — скорее всего, их нет. Можем пропуски заменить на 0.

```python
data['balcony'] = data['balcony'].fillna(0) #заполним пропуски в столбце 'balcony' нулями
```

##### Замена пропусков в столбце 'locality_name'

Найдем уникальные значения столбца locality_name.

```python
data['locality_name'].unique() #получим уникальные значения
```

Видим, что в наименованиях населенных пунктов есть буквы "е" и "ё". Это приводит к неявным дубликатам.

Заменим букву "ё" на "е".

```python
data['locality_name'] = data['locality_name'].str.replace('ё', 'е') #заменим буквы
```

По картографическим данным радиус г.Санкт-Петербург примерно 15 км. Заменим название населенного пункта на 'Санкт-Петербург', если расстояние до центра не превышает 15000м. Оставшиеся пропуски заменим на 'unknown'.

```python
data.loc[data['city_centers_nearest'] < 15000 , 'locality_name'].fillna('Санкт-Петербург', inplace=True) #заполним пропуски в столбце 'locality_name' на значение 'Санкт-Петербург' по условию - отдаленность от центра не более 15км
data['locality_name'] = data['locality_name'].fillna('unknown') #заменим пропуски на 'unknown'
```

В наименованиях содержатся значения одинаковые по смыслу, но разные по написанию, н-р, "поселок городского типа" и "поселок", что также приводит к дубликатам.

Для этого напишем функцию, которая с помощью регулярных выражений, заменит слова, начинающиеся с нижнего регистра, на пустое значение, оставив при этом слова только с заглавной буквы.

```python
import re #импортируем модуль re

#функция замены слов
def  get_proper_name(row):
    return re.sub(r'\b[а-я]+\s+', "", row['locality_name']) #заменим слова нижнего регистра с пробелом на пустое значение

data['locality_name'] = data.apply(get_proper_name, axis=1) #применим функцию к столбцу с наименованиями
```

##### Замена пропусков в столбце 'floors_total'

Можем предположить, что на многоэтажность дома влияет местоположение. Т.к. в зависимости от района строят характерные дома.

Заменим пропуски медианой, сгруппировав данные по наименованию населенного пункта и удаленности от центра.

```python
data['floors_total'] = data['floors_total'].fillna(data.groupby(['locality_name', 'city_centers_nearest'])['floors_total'].transform('median')) #заменим пропуски медианой, сгруппировав данные по столбцам 'city_centers_nearest', 'locality_name'
```

##### Замена пропусков в столбце 'living_area'

Для замены пропусков столбца 'living_area' найдем среднее соотношение жилой площади к общей и умножим на известное значение общей площади. Заполним пропуски этими расчетами.

```python
data['living_area'] = data['living_area'].fillna((data['living_area'] / data['total_area']).mean() * data['total_area']) #заменим пропуски результатом умножения значений столбца 'total_area' на среднее соотношение 'living_area' к 'total_area'
```

После заполнения пропусков, произведем подсчет заново.

```python
print(data.isna().sum())# подсчёт пропусков
```

### Изменение типов данных

В столбцах total_area, last_price, balcony, floors_total, city_centers_nearest, days_exposition заменим тип float64 на int64, предварительно округлив до целых.

```python
data['total_area'] = data['total_area'].round().astype('Int64') #заменим тип float64 на int64
```

```python
data['last_price'] = data['last_price'].astype('Int64') #заменим тип float64 на int64
```

```python
data['balcony'] = data['balcony'].astype('Int64') #заменим тип float64 на int64
```

```python
data['floors_total'] = data['floors_total'].round().astype('Int64') #округлим до целых и заменим тип float64 на int64
```

```python
data['city_centers_nearest'] = data['city_centers_nearest'].astype('Int64') #заменим тип float64 на int64
```

Высоту потолка округлим до десятых.

### Расчет дня недели, месяца и года публикации объявления

С помощью свойства dt получим в виде числа день недели (0-понедельник, 6-воскресенье), месяц(1-12), год.

```python
data['weekday'] = data['first_day_exposition'].dt.weekday #методом weekday получим день недели
data['month'] = data['first_day_exposition'].dt.month #методом month получим месяц
data['year'] = data['first_day_exposition'].dt.year #методом year получим год
```

### Расчет этажа квартиры; варианты — первый, последний, другой

Напишем функцию для определения категории этажа квартиры - первый, последний, другой.

```python
#функция для определения категории этажа
def get_floor(row):
    floors_total = row['floors_total']
    floor = row['floor']
    if floor == 1:
        return 1 #'первый'
    elif floor == floors_total:
        return 2 #'последний'
    else:
        return 3 #'другой'
    
data['floor_group'] = data.apply(get_floor, axis=1) #применим функцию к столбцу floors_total
```
